```python
import pandas as pd
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
import pickle
```

```python
df = pd.read_csv('Details.csv')
```

```python
df = df.sort_values(by='Test_accuracy',ascending=False)
```

```python

fig, axes = plt.subplots(2,1,figsize=(7,10))
sns.barplot(data=df,x='Train_accuracy',y='Model',palette='Set2',ax =axes[0])
sns.barplot(data=df,x='Train_accuracy',y='Model',palette='Set2',ax =axes[1])
plt.title("Accuracy Analysis")
fig.tight_layout()
fig.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the matrix
matrix = np.load('matrix.npy')
model = list(report_df.columns)

report  = report_df.iloc[0,:]

# Create a subplot for each matrix
fig, axes = plt.subplots(len(matrix), 1, figsize=(10, len(matrix) * 4))

# If there's only one matrix, wrap axes into a list for uniform handling
if len(matrix) == 1:
    axes = [axes]

# Iterate through the matrices and plot heatmaps
for i in range(len(matrix)):
    model_name = model[i]
    print("Classification Report : ")
    print(model_name)
    print(report[i]) 
    print("Confusion Matrix Analysis :\n",matrix[i])
    sns.heatmap(matrix[i], ax=axes[i], annot=True, cmap="viridis",
                xticklabels=['NO','YES'],yticklabels=['NO','YES']
                )
    axes[i].set_title(f"{model_name} ")

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```
