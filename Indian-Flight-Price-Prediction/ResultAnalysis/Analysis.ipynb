```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
results = pd.read_csv("Result.csv")
```

```python
results
```

```python
# sort the df
results = results.sort_values(by=['Test_score'],ascending=False)
```

```python
results.info()
```

```python
sns.set_style("dark") 
# chose best model 
fig, axes = plt.subplots(2, 1, figsize=(10, 18))

# Train Score Analysis
sns.barplot(data=results, y="Model", x="Train_score", palette="tab10", ax=axes[0])
axes[0].set_title("Train Score Analysis")
axes[0].set_xlabel("Train Score")
axes[0].set_ylabel("Model")

# Test Score Analysis
sns.barplot(data=results, y="Model", x="Test_score", palette="tab10", ax=axes[1])
axes[1].set_title("Test Score Analysis")
axes[1].set_xlabel("Test Score")
axes[1].set_ylabel("Model")

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```

```python
sns.set_style("dark") 

fig, axes = plt.subplots(2, 1, figsize=(10, 18))

# Train loss Analysis
sns.barplot(data=results, y="Model", x="Train_loss", palette="tab10", ax=axes[0])
axes[0].set_title("Train Score Analysis")
axes[0].set_xlabel("Train Loss")
axes[0].set_ylabel("Model")

# Test Loss Analysis
sns.barplot(data=results, y="Model", x="Test_loss", palette="tab10", ax=axes[1])
axes[1].set_title("Test Loss Analysis")
axes[1].set_xlabel("Test Loss")
axes[1].set_ylabel("Model")

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```
